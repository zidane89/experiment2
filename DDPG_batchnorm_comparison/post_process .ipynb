```python
import matplotlib.pyplot as plt 
import pickle 
import numpy as np 
```

```python
with open("DDPG6_1.pkl", "rb") as f: 
    data_6_1 = pickle.load(f)

with open("DDPG6_2.pkl", "rb") as f: 
    data_6_2 = pickle.load(f)

with open("DDPG6_3.pkl", "rb") as f: 
    data_6_3 = pickle.load(f)

with open("DDPG6_4.pkl", "rb") as f: 
    data_6_4 = pickle.load(f)
```

```python
plt.figure(figsize=(10, 7))
plt.plot(data_6_1["rewards"], label="[power_norm, SOC - 0.6, j_min, j_max]", linewidth=2)
plt.plot(data_6_2["rewards"], label="[power, SOC - 0.6, j_min, j_max]_norm", linewidth=2)
plt.plot(data_6_3["rewards"], label="[acc_norm, spd_norm, SOC - 0.6, j_min, j_max]", linewidth=2)
plt.plot(data_6_4["rewards"], label="[acc, spd, SOC - 0.6, j_min, j_max]_norm", linewidth=2)
plt.xlabel("episodes", fontsize=14)
plt.ylabel("episodic rewards", fontsize=14)
plt.legend(fontsize=14)
plt.grid()
plt.savefig("results.png")
```
