```python
import matplotlib.pyplot as plt 
import pickle 
import numpy as np 
```

```python
with open("DDPG1.pkl", "rb") as f: 
    data1 = pickle.load(f)[1]

with open("DDPG2.pkl", "rb") as f: 
    data2 = pickle.load(f)[1]

with open("DDPG3.pkl", "rb") as f: 
    data3 = pickle.load(f)[1]
```

```python
plt.figure(figsize=(10, 7))
plt.plot(data1["rewards"], label="", linewidth=2)
plt.plot(data2["rewards"], label="[power, SOC - 0.6, j_min, j_max]_norm", linewidth=2)
plt.plot(data3["rewards"], label="[acc_norm, spd_norm, SOC - 0.6, j_min, j_max]", linewidth=2)
# plt.plot(data_6_4["rewards"], label="[acc, spd, SOC - 0.6, j_min, j_max]_norm", linewidth=2)
plt.xlabel("episodes", fontsize=14)
plt.ylabel("episodic rewards", fontsize=14)
plt.legend(fontsize=14)
plt.grid()
plt.savefig("results.png")
```
