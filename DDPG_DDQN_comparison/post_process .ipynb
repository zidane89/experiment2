```python
import matplotlib.pyplot as plt 
import pickle 
```

```python
with open("DDPG.pkl", "rb") as f: 
    data_DDPG = pickle.load(f)[10]

with open("DDQN.pkl", "rb") as f: 
    data_DDQN = pickle.load(f)[10]
    
```

```python
plt.figure(figsize=(10, 7))
plt.plot(data_DDPG["rewards"], label="episodic_rewards_DDPG", linewidth=2)
plt.plot(data_DDQN["rewards"], label="episodic_rewards_DDQN", linewidth=2)
plt.xlabel("episodes", fontsize=14)
plt.ylabel("episodic rewards", fontsize=14)
plt.legend(fontsize=14)
plt.xlim([0, 150])
plt.grid()
plt.savefig("Comparison between DDQN and DDPG.png")
```
